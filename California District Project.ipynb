Below is an exploration of the district by distric mean property value in California, USA. The data are preprocessed, replacing incomplete data with the median value for that feature. The data is further standardised to remove the mean and set the variance to 1. The categorial feature in this dataset is inserted into the model via  1-hot encoding. New features are crafted and prove predictive, namely the number of bedrooms per house, the number of inhabitants per room and the number of rooms per household. Several machine learning algorithms are selected for the prediction of the median house value in an unknown, test set. Care was taken to ensure the same ratio of income segments was seen in the train set as in the full dataset, given this is one of the main predictors. Cross validation is performed. A grid search is used to find the most appropriate parameters of the models trained. Transformers are written in order to facilitate the addition of this code to a production pipeline. Random forests perform best on the test set. Most of the ideas were taken from the Aurelien Geron book called "Hands On Machine Learning with Sklearn, Keras and Tensorflow".

```python
import sklearn
import os
import urllib
import tarfile
import pandas as pd

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets","housing")

HOUSING_URL = DOWNLOAD_ROOT + HOUSING_PATH + "/" + "housing.tgz"
def save_from_url(housing_url=HOUSING_URL,housing_path=HOUSING_PATH):
        os.makedirs(housing_path,exist_ok=True)
        tgz_path = os.path.join(housing_path,"housing.tgz")
        urllib.request.urlretrieve(housing_url,tgz_path)
        housing_tgz = tarfile.open(tgz_path)
        housing_tgz.extractall(path=housing_path)
        housing_tgz.close()
```

```python
import numpy as np

def split_train_test(data,test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size=int(len(data)*test_ratio)
    test_indices = shuffled_indices[0:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
#define a hash table for unique IDs of data points
from zlib import crc32
```

```python
def test_set_check(identifier,test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff <test_ratio* 2**32
```

```python
def split_train_test_by_id(data,test_ratio,id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_:test_set_check(id_,test_ratio))
    return data.loc[~in_test_set],data.loc[in_test_set]
```

```python
from sklearn.base import BaseEstimator,TransformerMixin

rooms_ix,bedrooms_ix,population_ix, household_ix = 3,4,5,6

class CombinedAttribbutesAdder(BaseEstimator,TransformerMixin):
    def __init__(self,add_bedrooms_per_room=True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
        
    def fit(self,y=None):
        return self
    def transform(self,X):
        rooms_per_household = X[:,rooms_ix]/X[:,household_ix]
        population_per_household = X[:,population_ix]/X[:,household_ix]
        bedrooms_per_room = X[:,bedrooms_ix]/X[:,rooms_ix]
        if self.add_bedrooms_per_room:
            return np.c_[X,rooms_per_household,population_per_household,bedrooms_per_room]
        else:
            return np.c_[X,rooms_per_household,population_per_household]
        
        attr_adder = CombinedAttribbutesAdder(add_bedrooms_per_room=False)
        housing_extra_attribs = attr_adder.transform(housing.values)
        
        
```

```python
from sklearn.ensemble import RandomForestRegressor
class FeatureSelector(BaseEstimator,TransformerMixin):
    def __init__(self,attributes):
        self.attributes = attributes
        return None
    def fit(self,X,y):
        model_init= RandomForestRegressor()
        model = model_init.fit(X,y)
        self.fitted = model
        return self
    def transform(self,X):
        feature_rank = self.fitted.feature_importances_
        feature_list = [sorted(zip(feature_rank,self.attributes),reverse=True)]
      
        #print(feature_list)
        indices_ordered=np.argsort(-abs(feature_rank)) # need from max to min
        labels_selected=[]
        for x in indices_ordered[0:10]:
            labels_selected.append(self.attributes[x])

        return X[:,indices_ordered[0:10]],labels_selected#,self.attributes[indices_ordered[0:10]]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')), ('attribs_adder',CombinedAttribbutesAdder()),('standard_scaler',StandardScaler())]) 

housing_num_tr = num_pipeline.fit_transform(housing_num)
```
